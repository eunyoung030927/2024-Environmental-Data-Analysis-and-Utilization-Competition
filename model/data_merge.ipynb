```python
air_22_1 = pd.read_csv(r"C:\Users\dnjsw\Desktop\데이터 분석 개인 프젝\2024 환경 데이터 분석 공모전\data\대기질데이터_전처리\4_2022pm_1_6.csv").drop(columns = ['Unnamed: 0.1','Unnamed: 0'])
air_22_2 = pd.read_csv(r"C:\Users\dnjsw\Desktop\데이터 분석 개인 프젝\2024 환경 데이터 분석 공모전\data\대기질데이터_전처리\1_2022pm_7_12_.csv").drop(columns = ['Unnamed: 0.1','Unnamed: 0'])
air_23_1 = pd.read_csv(r"C:\Users\dnjsw\Desktop\데이터 분석 개인 프젝\2024 환경 데이터 분석 공모전\data\대기질데이터_전처리\2_2023pm_1_6.csv").drop(columns = ['Unnamed: 0.1','Unnamed: 0'])
air_23_2 = pd.read_csv(r"C:\Users\dnjsw\Desktop\데이터 분석 개인 프젝\2024 환경 데이터 분석 공모전\data\대기질데이터_전처리\3_2023pm_7_12.csv").drop(columns = ['Unnamed: 0.1','Unnamed: 0'])
```

```python
dfs = [air_22_1, air_22_2, air_23_1, air_23_2]

# 빈 리스트를 만들어 각 데이터프레임을 담을 것입니다.
merged_dfs = []

for df in dfs:
    # '지역' column을 공백을 기준으로 분할하여 새로운 데이터프레임 생성
    new_columns = df['지역'].str.split(' ', n=1, expand=True)

    # 새로운 데이터프레임에 컬럼명 추가
    new_columns.columns = ['지역', '지역명']

    # 기존 데이터프레임과 새로운 데이터프레임을 합침
    merged_df = pd.concat([df.drop(columns='지역'), new_columns], axis=1)
    
    # 최종 데이터프레임 리스트에 추가
    merged_dfs.append(merged_df)

# 모든 데이터프레임을 위아래로 병합
final_df = pd.concat(merged_dfs, ignore_index=True)
final_df.head()
```
