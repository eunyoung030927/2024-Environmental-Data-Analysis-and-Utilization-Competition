```python
from datetime import datetime

# df[['year', 'month', 'day']]=df[['year', 'month', 'day']].astype('string')
def combine_to_yyyymmdd(row):
    date = datetime(row['year'], row['month'], row['day'])
    return date.strftime('%Y%m%d')

# 변환 적용
df['timestamp'] = df.apply(combine_to_yyyymmdd, axis=1).astype('int64')

df['timestamp']
```

```python
# df.set_index('timestamp', inplace=True) <- 이거 맨 나중에
df.drop(columns=['year', 'month', 'day', 'Unnamed: 0'], inplace=True)
df
```

우리는 2차선 이하만 보면 되니까 그것들만 보자

```python
lanes = lanes[lanes['차선'] <= 2]
```

```python
lanes['차선'].unique()
```

```python
for _, lane in lanes.iterrows():
    
    print(lane['도로명'])
```

```python
df['차선'] = '0'
for _, lane in lanes.iterrows():
     lane_name = lane['도로명']
     lane_loc = lane['지역명']
     df.loc[df['도로명'].str.contains(lane_name) & df['지역명'].str.contains(lane_loc), '차선'] = lane['차선']
df['차선']
```

```python
df = df[df['차선'] != '0'] 
df
```
