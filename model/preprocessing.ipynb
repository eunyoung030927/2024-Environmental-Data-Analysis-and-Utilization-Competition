```python
from datetime import datetime

# df[['year', 'month', 'day']]=df[['year', 'month', 'day']].astype('string')
def combine_to_yyyymmdd(row):
    date = datetime(row['year'], row['month'], row['day'])
    return date.strftime('%Y%m%d')

# 변환 적용
df['timestamp'] = df.apply(combine_to_yyyymmdd, axis=1).astype('int64')

df['timestamp']
```

```python
final_df = pd.read_csv('../data/환경 공모전 최종 데이터.csv')
final_df
```

```python
final_df.columns
```

```python
lspd = pd.read_csv('../data/차선정보_전처리/lanes_speed.csv')
lspd
```

```python
merged_speed = pd.merge(spdf, lspd[['도로명', '최고속도']], on='도로명', how='inner').dropna(axis=1)
merged_speed
```

```python
merged_speed.isnull().sum()
```

```python
merged_speed = merged_speed.groupby('도로명').mean().reset_index().drop(['기준월'], axis=1)
```

```python
merged_speed
```

```python
final_df = final_df[~final_df['도로명'].str.contains('→')]
final_df
```

```python
final_df['평균속도'] = np.nan
final_df['최고속도'] = np.nan

for idx, row in final_df.iterrows():
    road_name = row['도로명']
    if road_name in merged_speed['도로명'].values: 
        final_df.loc[idx, '평균속도'] = merged_speed[merged_speed['도로명'] == road_name]['평균속도'].values
        final_df.loc[idx, '최고속도'] = merged_speed[merged_speed['도로명'] == road_name]['최고속도'].values

        print(merged_speed[merged_speed['도로명'] == road_name]['평균속도'].values, merged_speed[merged_speed['도로명'] == road_name]['최고속도'].values)

final_df
```

```python
final_df[final_df.isnull().any(axis=1)]['지역명'].unique()
```

```python
for idx, df in final_df[final_df[['평균속도', '최고속도']].isnull().any(axis=1)].iterrows():
    location = df['지역명']
    avgspd = final_df[final_df['지역명']==location]['평균속도'].mean()
    highspd = final_df[final_df['지역명']==location]['최고속도'].mean()
    
    final_df.loc[idx, '평균속도'] = avgspd
    final_df.loc[idx, '최고속도'] = highspd
```

```python
fdf = final_df.drop('Unnamed: 0', axis=1)
fdf
```
